```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create Plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Annotate
# Create the plot and add a label for the legend.
plt.plot(x_axis,y_axis,marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0,45)

# Create a title.
plt.title("Pyber Fare by Month")

#add the legend
plt.legend()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis,marker="*", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes.
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')

ax.set_title("Title")

#Add legend
ax.legend("Pyber Fare by Month")

#Add grid
ax.grid()

# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create Scatter plot using plt.Plot() method 
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create Scatter plot using plt.Plot() method 
plt.scatter(y_axis,x_axis, c="red", label="Chicago")

#add legend
plt.legend()

#add title
plt.title("PyBer Fare by Month")

#add label:
plt.xlabel("Fare ($)")
plt.ylabel("Date")

#Swicth for Jan to be at top
plt.gca().invert_yaxis()
```

```python
#Bubble charts using MATLAB 
plt.scatter(x_axis,y_axis, s=y_axis)
```

```python
#bigger the marker size
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#refactoring the above code using list comprehension
plt.scatter(x_axis,y_axis, s=[i*3 for i in y_axis])
```

```python
#Scatter plots using object-oriented method
fig, ax=plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
#Bubbl plots using object-oriented method
fig, ax=plt.subplots()
ax.scatter(y_axis,x_axis, color="skyblue", s=[i*5 for i in y_axis], label="Boston", alpha=0.8, edgecolors='black')


ax.set_title("PyBer Fare by Month")
ax.legend()
ax.set_xlim(0,50)
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
plt.gca().invert_yaxis()
```

```python
#Creating Piechart with MatLab method
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
#changing the pie chart size
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, colors=colors,labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
#Creating Piechart with Object Oriented Interface
#changing the pie chart size
explode=[0,0,0.2,0,0,0,0.2,0,0,0,0,0]
fig, ax=plt.subplots(figsize=(8, 8))
ax.pie(y_axis,colors=colors,
       labels=x_axis,explode=explode,
       startangle=90,counterclock=False,
       shadow=True,autopct='%.1f%%')

ax.legend(title="Date")
plt.show()
```
